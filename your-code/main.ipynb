## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
uniform.rvs(2, 8, size=10)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set seaborn style
sns.set(style="whitegrid")

# Function to generate uniformly distributed random numbers
def generate_uniform_random(bottom, ceiling, count):
    return np.random.uniform(low=bottom, high=ceiling, size=count)

# Generate data
uniform_data_1 = generate_uniform_random(10, 15, 100)
uniform_data_2 = generate_uniform_random(10, 60, 1000)

# Create output directory
output_dir = "/mnt/data"
os.makedirs(output_dir, exist_ok=True)




```

```python
# Plot histogram for first dataset
plt.figure(figsize=(10, 6))
plt.hist(uniform_data_1, bins=10, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution: bottom=10, ceiling=15, count=100')
plt.xlabel('Value')
plt.ylabel('Count')
plt.tight_layout()
plot1_path = os.path.join(output_dir, "uniform_distribution_10_15_100.png")
#plt.savefig(plot1_path)
#plt.close()
plt.show()
```

```python
# Plot histogram for second dataset
plt.figure(figsize=(10, 6))
plt.hist(uniform_data_2, bins=10, color='salmon', edgecolor='black')
plt.title('Uniform Distribution: bottom=10, ceiling=60, count=1000')
plt.xlabel('Value')
plt.ylabel('Count')
plt.tight_layout()
plot2_path = os.path.join(output_dir, "uniform_distribution_10_60_1000.png")
#plt.savefig(plot2_path)
#plt.close()
plt.show()
print("Plots saved:", plot1_path, plot2_path)
```

How are the two distributions different?

```python
The first histogram will show more variability due to the smaller sample size and narrower range.

The second histogram will appear flatter and more evenly distributed, thanks to the larger sample size and wider range.
```

```python
from scipy.stats import norm 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Function to generate normally distributed random numbers
def generate_normal_data(mean, std, size):
    return np.random.normal(loc=mean, scale=std, size=size)





```

```python
# Generate datasets
data_std1 = generate_normal_data(mean=10, std=1, size=1000)
data_std50 = generate_normal_data(mean=10, std=50, size=1000)
```

```python
# Plot histograms
plt.figure(figsize=(10, 6))
sns.histplot(data_std1, color='blue', label='Mean=10, Std=1', kde=True, stat="density", bins=30)
sns.histplot(data_std50, color='red', label='Mean=10, Std=50', kde=True, stat="density", bins=30)
plt.title('Comparison of Normal Distributions with Different Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# Function to generate exponentially distributed numbers with mean 10
def generate_exponential_data(size, mean=10):
    scale = mean  # scale parameter for exponential distribution is equal to the mean
    return np.random.exponential(scale=scale, size=size)

# Generate data
data_10 = generate_exponential_data(10)
data_100 = generate_exponential_data(100)

# Create output directory if it doesn't exist
# output_dir = "/mnt/data"
# os.makedirs(output_dir, exist_ok=True)

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for size 10
sns.histplot(data_10, bins=100, kde=False, ax=axes[0], color='skyblue')
axes[0].set_title('Exponential Distribution (size=10, mean=10)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Histogram for size 100
sns.histplot(data_100, bins=100, kde=False, ax=axes[1], color='salmon')
axes[1].set_title('Exponential Distribution (size=100, mean=10)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
plt.show()

# Save the plot
# plot_path = os.path.join(output_dir, "exponential_distributions.png")
# plt.tight_layout()
# plt.savefig(plot_path)
# plt.close()

print("Histograms saved as 'exponential_distributions.png'")
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
